```python
# ויזואליזציות
import pandas as pd
import numpy as np
import matplotlib as mpl
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
pok = pd.read_csv('pokemon.csv')
ct1 = pd.crosstab(pok['generation'],pok['is_legendary'],normalize='index') # קודה שעושה קרוס בין שני העמודות, כאן היא סוכמת כמה אגדיים יש בכל דור 1-7 
                                                                           # הנורמלייז ממיר את הנתונים לעשרוניים שכל שורה סכומה הוא אחד, מעין אחוז פוקימונים מכל סוג בכל דור
# יצירת גרף שמראה את היחס בכל דור ודור בין אגדיים לרגילים
# זוהי ויזואליזציה חד מימדית
ct1.plot(kind='bar',figsize=(5,2))
plt.title('Legendary vs. normal Pokemons',fontsize=10)
plt.xlabel('Generation')
plt.ylabel('Relative frequency')
```
