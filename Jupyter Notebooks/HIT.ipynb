```python
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib as mpl
%matplotlib inline
df = pd.read_csv("Bike_sharing_data.csv")
plt.scatter(df.registered,df.cnt)
df.cnt[df.cnt>5000].count() # חישוב כמה ימים(רשומות) הייתה צריכה ביותר מ5000 אופניים
df.cnt[df.cnt>5000].count()/df.cnt.count() # חישוב ההסתברות של כמה ימים (רשומות) הייתה צריכה ביותר מ5000 אופניים
```
